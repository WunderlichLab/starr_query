```python
import pandas as pd
import re
import numpy as np
```

```python
# Read the files
hksm_df = pd.read_csv('base_data/EG_HKSM_df_11_6.csv', header = 0)
e20_df = pd.read_csv('base_data/EG_20E_tc_04_22.csv', header = 0)
con_df = pd.read_csv('base_data/EG_Con_tc_04_22.csv', header = 0)
```

```python
hksm_df['Origin'] = 'HKSM'
tf_col = ['kay_Jra','crp','bergman_Rel','GATA_elemento','srp_SANGER','xrp1','XBP1','EcR','ERR','USP','da','slp2_forkhead','Hnf4','Eip74EF']
hksm_df['tf_counts'] = hksm_df[tf_col].apply(lambda row: '; '.
                                             join([f'{col}: {int(row[col])}' for col in tf_col if row[col] > 0]),
                                             axis = 1)
hksm_df.rename(columns = {'Total_binding_sites ': 'TBS'}, inplace = True)
hksm_df
```

```python
e20_df['Origin'] = 'E20'
tf_col = ['XBP1','Trl','xrp1','USP','bergman_EcR_usp','CF2','bergman_Rel','da','crp','EcR','GATA_elemento','ERR','Eip74EF','h','gcm','kay_Jra','Hnf4','srp_SANGER','slp2_forkhead','Rel_FFS','Tgo']
e20_df['tf_counts'] = e20_df[tf_col].apply(lambda row: '; '.
                                           join([f'{col}: {int(row[col])}' for col in tf_col if row[col] > 0]),
                                           axis = 1)
e20_df
```

```python
con_df['Origin'] = 'Control'
tf_col = ['XBP1','Trl','xrp1','USP','bergman_EcR_usp','CF2','bergman_Rel','da','crp','EcR','GATA_elemento','ERR','Eip74EF','h','gcm','kay_Jra','Hnf4','srp_SANGER','slp2_forkhead','Rel_FFS','Tgo']
con_df['tf_counts'] = con_df[tf_col].apply(lambda row: '; '.
                                           join([f'{col}: {int(row[col])}' for col in tf_col if row[col] > 0]),
                                           axis = 1)
con_df.drop('FB_Gene',axis = 1, inplace = True)
con_df

# Checked for mismatches where both fields are not null manually and found zero columns. Hence dropped FB_gene
```

```python
# Combine the dataframes
enhancer_df = pd.concat([hksm_df, e20_df, con_df])
enhancer_df.rename(columns = {'tf_counts': 'TF Counts'},inplace = True)
enhancer_df
```

# Enhancer Table

```python
# Extract the 'Enhancer' column
enhancer = enhancer_df['Enhancer']

# Parse chromosome, start, and end from the Enhancer string
enhancer_df['Chromosome'] = enhancer.apply(lambda x: x.split(':')[0])
enhancer_df['Start'] = enhancer.apply(lambda x: int(re.search(r':(\d+)-', x).group(1)))
enhancer_df['End'] = enhancer.apply(lambda x: int(re.search(r'-(\d+)', x).group(1)))
```

```python
# Display the updated DataFrame and store it to csv
enhancer_df.reset_index(drop = True, inplace = True)
enhancer_df[['Enhancer', 'Chromosome', 'Start', 'End','TF Counts','TBS']].drop_duplicates().to_csv('enhancer.csv', index = False)
```

# Genes Table

```python
# gene csv
gene_df = enhancer_df[['Genes','Immune Process','Time_cluster']]
gene_df.reset_index(drop = True, inplace = True)
gene_df.rename(columns = {'Genes':'gene_id'},inplace = True)
gene_df
```

```python
# Make a unique list of genes
unique_genes =gene_df['gene_id'].dropna().unique()
unique_genes = unique_genes.tolist()
```

```python
# Read gtf file
gtf_file = pd.read_csv('dmel-all-r6.62.gtf',header = None, sep = '\t')
gtf_file = gtf_file.loc[(gtf_file[2] == 'gene')]

# Regex
gtf_file['gene_id'] = gtf_file[8].str.extract(r'gene_id\s+"(FBgn\d+)"')
gtf_file['gene_name'] = gtf_file[8].str.extract(r'gene_symbol\s+"([^"]+)"')
gtf_file
```

```python
gene_df
```

```python
# Filter
filtered_genes = gtf_file[gtf_file['gene_id'].isin(unique_genes)]
filtered_genes

result_df = pd.merge(filtered_genes[['gene_id', 0, 3, 4, 'gene_name']],gene_df[['gene_id','Immune Process', 'Time_cluster']],on = 'gene_id', how = 'left')
result_df.columns = ['GeneID','Chromosome', 'Start', 'End', 'GeneName','Immune Process','Time_cluster']
result_df.drop_duplicates(ignore_index = True,inplace = True)
result_df.reset_index(drop=True, inplace=True)
result_df
```

```python
result_df.to_csv('genes.csv', index = False)
```
